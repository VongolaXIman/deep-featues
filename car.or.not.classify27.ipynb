```python
import graphlab
import numpy as np
```

```python
sf = graphlab.SFrame.read_json('car.or.not.train.json')
sf.head()
```

```python
sf.shape
```

```python
sf['vgg16_features'].head(1) # dont know why this column is not array but list when i read json to SFrame
```

```python
sf['vgg16_features2'] = np.asarray(sf['vgg16_features']) # have to convert this column list to array
sf['vgg16_features2'].head(1)
```

```python
sf.head()
```

```python
vgg16_logi_model = graphlab.logistic_classifier.create(sf,
                                                         features=['vgg16_features2'],
                                                         target='label_car2')
```

```python
vgg16_logi_model2 = graphlab.logistic_classifier.create(sf,
                                                         features=['vgg16_features2'],
                                                         target='label_car')
```

```python
sf2 = graphlab.SFrame.read_json('car.or.not.test.json')
sf2.head()
```

```python
sf2['vgg16_features2'] = np.asarray(sf2['vgg16_features']) # have to convert this column list to array
sf2['vgg16_features2'].head(1)
```

```python
vgg16_logi_model.evaluate(sf2)
```

```python
vgg16_logi_model2.evaluate(sf2)
```
