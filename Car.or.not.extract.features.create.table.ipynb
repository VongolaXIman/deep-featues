```python
df = pd.DataFrame([{'id':'1', 'label_car':1, 'array':0, 'vgg16_features':0}])

df
```

```python
import os
model = VGG16(weights='imagenet', include_top=False)
for filename in os.listdir('publish_dataset/train/positive'):
    image_array16 = np.array(Image.open(os.path.join('publish_dataset/train/positive/', filename)))
    img16 = image.load_img(os.path.join('publish_dataset/train/positive/', filename), target_size=(224, 224)) # resize to 224*224 
    x16 = image.img_to_array(img16)
    x16 = np.expand_dims(x16, axis=0)
    x16 = preprocess_input(x16)
    features16= model.predict(x16)
    features16=features16.flatten()
    print(filename)
    df = df.append(pd.DataFrame([{'id':filename, 'label_car':1, 'array':image_array16, 'vgg16_features':features16}]))

for filename in os.listdir('publish_dataset/train/negative'):
    image_array16 = np.array(Image.open(os.path.join('publish_dataset/train/negative/', filename)))
    img16 = image.load_img(os.path.join('publish_dataset/train/negative/', filename), target_size=(224, 224)) # resize to 224*224 
    x16 = image.img_to_array(img16)
    x16 = np.expand_dims(x16, axis=0)
    x16 = preprocess_input(x16)
    features16= model.predict(x16)
    features16=features16.flatten()
    print(filename)
    df = df.append(pd.DataFrame([{'id':filename, 'label_car':0, 'array':image_array16, 'vgg16_features':features16}]))  

df.head()
```
